```python
# Setting the sensor generated "time" column to an appropriate timestamp type 

for i in gc_list:
    i['%time'] = pd.to_datetime(i['%time'], unit='d', origin=pd.Timestamp('1900-01-01'))

team_6_gc.head() # checking if the conversion was successful 
```

```python
# Set the '%time' column as the DataFrames' index
for i in gc_list:
    i.set_index('%time', inplace=True)
```

```python
# Group the data by week and calculate the average of the 'Tair' for smoothening the line graph
weekly_average_1 = team_1_gc['Tair'].resample('W').mean()
weekly_average_2 = team_2_gc['Tair'].resample('W').mean()
weekly_average_3 = team_3_gc['Tair'].resample('W').mean()
weekly_average_4 = team_4_gc['Tair'].resample('W').mean()
weekly_average_5 = team_5_gc['Tair'].resample('W').mean()
weekly_average_6 = team_6_gc['Tair'].resample('W').mean()

# checking if the resampling was successful
weekly_average_1.head()
```

```python
# We are firstly creating two function in order to create later the line plots much easier and faster

# The "plot1" function has the y_axis ticks parameters inside something that the "plot2" function doesn't have
def plot1(a, b, c, d, e, f, title, ylabel, tick1, tick2, tick3):
    fig, ax = plt.subplots()
    
    ax.plot(a, color='forestgreen', label ='Team_1')
    ax.plot(b, color='olive', label ='Team_2')
    ax.plot(c, color='dimgray', label ='Team_3')
    ax.plot(d, color='firebrick', label ='Team_4')
    ax.plot(e, color='purple', label ='Team_5')
    ax.plot(f, color='orange', label ='Team_6')
    
    ax.set(title=title,
           ylabel=ylabel,
           xlabel='Timeframe')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    ax.yaxis.set_ticks(np.arange(tick1, tick2, tick3))
    plt.grid()
    plt.show()

def plot2(a, b, c, d, e, f, title, ylabel):
    fig, ax = plt.subplots()
    
    ax.plot(a, color='forestgreen', label ='Team_1')
    ax.plot(b, color='olive', label ='Team_2')
    ax.plot(c, color='dimgray', label ='Team_3')
    ax.plot(d, color='firebrick', label ='Team_4')
    ax.plot(e, color='purple', label ='Team_5')
    ax.plot(f, color='orange', label ='Team_6')
    
    ax.set(title=title,
           ylabel=ylabel,
           xlabel='Timeframe')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    plt.grid()
    plt.show() 
```

```python
# Plot The weekly Average Temperature in each compartment

plot2(weekly_average_1, weekly_average_2, weekly_average_3, weekly_average_4, weekly_average_5, weekly_average_6, \
      'Weekly Avg Temperature in each compartment', 'Temperature [°C]')
```

**Figure 1** *Weekly average temperature (°C) in the six greenhouse compartments*

_Comments_
- Some compartments showed a relatively stable temperature throughout the growing season (Team 1, 3 and 4), but others showed higher temperatures during the first weeks, then a moderate level and a large increase at the end (Team 5 and 6), probably to accelerate development early in the season and fasten fruit ripening at the end of the season
- Team 6 is shown to reach high temperature regimes that is not being showed in the **Figure 2**, thing that happened probably by high humidity and limit ventilation when possible
- The reference growers (Team 3) applied relatively high temperatures throughout the season, which also resulted in the highest heating use (**Figure 2**)

**Figure 6** *Weekly average CO2 concentration (ppm) in each greenhouse compartment*

_Comments_
- Values vary between 600 and 900 ppm for most of the season, lowering to 400–600 ppm at the end of the season (> sun/temperature => > ventilation)
- Team 1 and 2 maintained high CO2 levels throughout the season. Team 1 achieved that with very high CO2 dosage (**Figure 7**), while Team 2 achieved the same levels with lower CO2 dosage(**Figure 7**)

```python
# We now want to find the Daily CO2 dosage (g/m2 day) in each greenhouse compartment. For that we have to convert (kg/m2 hour)
# to (kg/m2 5min), then group by Day and suming the 5 minute measurements, then convert (kg/m2 d) to (g/m2 d) and lastly for 
# smoothening the curve we use the "Simple Moving Average" method or in other words with the .rolling() method

co2_dosage_1 = (team_1_gc['co2_dos']/60*5).resample('D').sum().mul(1000).rolling(3, min_periods=1).mean() 
co2_dosage_2 = (team_2_gc['co2_dos']/60*5).resample('D').sum().mul(1000).rolling(3, min_periods=1).mean() 
co2_dosage_3 = (team_3_gc['co2_dos']/60*5).resample('D').sum().mul(1000).rolling(3, min_periods=1).mean()
co2_dosage_4 = (team_4_gc['co2_dos']/60*5).resample('D').sum().mul(1000).rolling(3, min_periods=1).mean()
co2_dosage_5 = (team_5_gc['co2_dos']/60*5).resample('D').sum().mul(1000).rolling(3, min_periods=1).mean()
co2_dosage_6 = (team_6_gc['co2_dos']/60*5).resample('D').sum().mul(1000).rolling(3, min_periods=1).mean()

co2_dosage_1.head(15)
```

```python
# Plot the Daily CO2 Dosage (g/m2) in each compartment

plot1(co2_dosage_1,co2_dosage_2,co2_dosage_3,co2_dosage_4,co2_dosage_5,co2_dosage_6,\
      'Daily CO2 Dosage [g/m2] in each compartment','CO2 [g/m2]',0,160,20)
```

```python
# Plot the Correlation between EC & TSS in the compartments

# Obtain m (slope) and b(intercept) of linear regression line (here we use the "merged_gc_qc" dataframe)
m, b = np.polyfit(merged_gc_qc['EC_drain_PC'], merged_gc_qc['TSS'], 1)

# Create correlation - Scatter plot between EC in Drain Water (dS/m) & TSS (°Brix)
fig, ax = plt.subplots()

ax.scatter(merged_gc_qc_1['EC_drain_PC'], merged_gc_qc_1['TSS'], color='blue', alpha=0.5)
ax.scatter(merged_gc_qc_2['EC_drain_PC'], merged_gc_qc_2['TSS'], color='blue', alpha=0.5)
ax.scatter(merged_gc_qc_3['EC_drain_PC'], merged_gc_qc_3['TSS'], color='blue', alpha=0.5)
ax.scatter(merged_gc_qc_4['EC_drain_PC'], merged_gc_qc_4['TSS'], color='blue', alpha=0.5)
ax.scatter(merged_gc_qc_5['EC_drain_PC'], merged_gc_qc_5['TSS'], color='blue', alpha=0.5)
ax.scatter(merged_gc_qc_6['EC_drain_PC'], merged_gc_qc_6['TSS'], color='blue', alpha=0.5)

ax.plot(merged_gc_qc['EC_drain_PC'], m*merged_gc_qc['EC_drain_PC']+b, color = 'black', 
        alpha=0.4, label = 'y={}*x+{}'.format(round(m,4),round(b,4)))

ax.set(title='Correlation between EC & TSS in the compartments', 
       ylabel='TSS [°]',
       xlabel='EC [dS/m]')
plt.legend(loc='center left', bbox_to_anchor=(0.55, 0.85),
          fancybox=True, shadow=True)
plt.grid()
plt.show()
```
